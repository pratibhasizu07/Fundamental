```python
# üßÆ Student Result Management System
# Developed for Google Colab using Python + Pandas

import pandas as pd

# Create an empty DataFrame
columns = ["Roll No", "Name", "Subject 1", "Subject 2", "Subject 3", "Total", "Average", "Grade"]
students = pd.DataFrame(columns=columns)

# Function to calculate grade
def calculate_grade(avg):
    if avg >= 90:
        return "A+"
    elif avg >= 80:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 60:
        return "C"
    elif avg >= 50:
        return "D"
    else:
        return "F"

# Function to add a student
def add_student(roll, name, s1, s2, s3):
    global students
    total = s1 + s2 + s3
    avg = total / 3
    grade = calculate_grade(avg)
    data = pd.DataFrame([[roll, name, s1, s2, s3, total, avg, grade]], columns=students.columns)
    students = pd.concat([students, data], ignore_index=True)
    print(f"\n‚úÖ Record added for {name} (Roll No: {roll})")

# Function to display all students
def display_students():
    if students.empty:
        print("\n‚ö†Ô∏è No records available.")
    else:
        print("\nüìã All Student Records:\n")
        display(students)

# Function to search student by roll number
def search_student(roll):
    record = students[students["Roll No"] == roll]
    if record.empty:
        print("\n‚ùå Student not found.")
    else:
        print("\nüîç Student Record:\n")
        display(record)

# Function to update student marks
def update_student(roll, s1, s2, s3):
    global students
    index = students[students["Roll No"] == roll].index
    if index.empty:
        print("\n‚ùå Student not found.")
    else:
        total = s1 + s2 + s3
        avg = total / 3
        grade = calculate_grade(avg)
        students.loc[index, ["Subject 1", "Subject 2", "Subject 3", "Total", "Average", "Grade"]] = [s1, s2, s3, total, avg, grade]
        print(f"\n‚úèÔ∏è Record updated for Roll No: {roll}")

# Function to delete student record
def delete_student(roll):
    global students
    index = students[students["Roll No"] == roll].index
    if index.empty:
        print("\n‚ùå Student not found.")
    else:
        students = students.drop(index)
        print(f"\nüóëÔ∏è Record deleted for Roll No: {roll}")

# --- Example Usage ---

# Adding students
add_student(1, "Aarav", 85, 92, 88)
add_student(2, "Priya", 76, 80, 79)
add_student(3, "Rohan", 95, 90, 93)

# Display all records
display_students()

# Search student
search_student(2)

# Update marks
update_student(2, 90, 91, 89)

# Display updated table
display_students()

# Delete a student
delete_student(1)

# Final result list
display_students()
```
